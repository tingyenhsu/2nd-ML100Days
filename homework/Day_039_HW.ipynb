2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?
##### 共現性高時w會變得比較大，而ridge regression會是在w的平方後再加上一個alpha係數，將這個數字也加進去懲罰項中。因此只要w變得大時，雖然模型看似擬合，為了避免過擬合或是受噪音影響過大的問題，懲罰項也會隨之變大，讓模型不要往w過大的方向走。
